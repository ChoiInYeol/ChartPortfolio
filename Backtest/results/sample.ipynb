```python
import pandas as pd

path = 'weights_CNN + TCN Top 50.csv'

data = pd.read_csv(path, index_col=0)

# 각 열의 합이 0인지 확인하여 0이 아닌 열만 선택
data = data.loc[:, (data != 0).any()]
data
```

```python
data.plot()
```
