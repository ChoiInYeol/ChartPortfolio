```python
import h5py
import pandas as pd
import numpy as np

def load_and_preprocess_data(file_path):
    """
    H5 파일에서 데이터를 로드하고 전처리하여 데이터프레임으로 변환합니다.
    
    Args:
        file_path (str): H5 파일의 경로
    
    Returns:
        tuple: (X_df, y_df) 형태의 튜플. X_df는 입력 데이터, y_df는 타겟 데이터의 데이터프레임
    """
    with h5py.File(file_path, 'r') as f:
        X = f['x'][:]
        y = f['y'][:]
    
    # X 데이터 전처리
    n_samples, n_timesteps, n_features = X.shape
    X_reshaped = X.reshape(n_samples * n_timesteps, n_features)
    
    feature_names = [f'feature_{i}' for i in range(n_features)]
    X_df = pd.DataFrame(X_reshaped, columns=feature_names)
    X_df['sample_id'] = np.repeat(np.arange(n_samples), n_timesteps)
    X_df['timestep'] = np.tile(np.arange(n_timesteps), n_samples)
    
    # y 데이터 전처리
    n_samples, n_future_steps, n_features = y.shape
    y_reshaped = y.reshape(n_samples * n_future_steps, n_features)
    
    y_df = pd.DataFrame(y_reshaped, columns=feature_names)
    y_df['sample_id'] = np.repeat(np.arange(n_samples), n_future_steps)
    y_df['future_step'] = np.tile(np.arange(n_future_steps), n_samples)
    
    return X_df, y_df

# 데이터 로드 및 전처리
X_df, y_df = load_and_preprocess_data('data/train_dataset.h5')

# 결과 확인
print("X 데이터프레임:")
print(X_df.head())
print("\ny 데이터프레임:")
print(y_df.head())

# 데이터 형태 확인
print("\nX 데이터프레임 형태:", X_df.shape)
print("y 데이터프레임 형태:", y_df.shape)
```
